```python
import pandas as pd
import numpy as np
from scipy import stats

df = pd.read_csv("ab_test_data.csv")

group_A = df[df['group'] == 'A']
group_B = df[df['group'] == 'B']

print("Group A size:", len(group_A))
print("Group B size:", len(group_B))
```

```python
conv_A = group_A['converted'].sum()
conv_B = group_B['converted'].sum()
n_A = len(group_A)
n_B = len(group_B)

expected_success_A = n_A * (conv_A/n_A)
expected_fail_A = n_A - expected_success_A
expected_success_B = n_B * (conv_B/n_B)
expected_fail_B = n_B - expected_success_B

print("Expected successes/failures:")
print(f"Group A: {expected_success_A:.0f} successes, {expected_fail_A:.0f} failures")
print(f"Group B: {expected_success_B:.0f} successes, {expected_fail_B:.0f} failures")

print("\nTime on Site:")
print("Group A: mean =", round(group_A['time_on_site'].mean(),2), 
      "SD =", round(group_A['time_on_site'].std(),2))
print("Group B: mean =", round(group_B['time_on_site'].mean(),2), 
      "SD =", round(group_B['time_on_site'].std(),2))

print("\nPages Viewed:")
print("Group A: mean =", round(group_A['pages_viewed'].mean(),2), 
      "SD =", round(group_A['pages_viewed'].std(),2))
print("Group B: mean =", round(group_B['pages_viewed'].mean(),2), 
      "SD =", round(group_B['pages_viewed'].std(),2))
```

```python
p_A = conv_A / n_A
p_B = conv_B / n_B
p_pool = (conv_A + conv_B)/(n_A + n_B)
se = np.sqrt(p_pool * (1 - p_pool) * (1/n_A + 1/n_B))
z_stat = (p_B - p_A)/se
p_value_conv = 1 - stats.norm.cdf(z_stat)  

print("Conversion Rate:")
print("p_A =", round(p_A,4), "p_B =", round(p_B,4))
print("Z-statistic =", round(z_stat,4))
print("One-tailed p-value =", round(p_value_conv,6))

t_time, p_time_two = stats.ttest_ind(group_B['time_on_site'],
                                     group_A['time_on_site'],
                                     equal_var=False)
p_time = p_time_two / 2  
print("\nTime on Site:")
print("Mean A =", round(group_A['time_on_site'].mean(),2),
      "Mean B =", round(group_B['time_on_site'].mean(),2))
print("T-statistic =", round(t_time,4))
print("One-tailed p-value =", round(p_time,6))

t_pages, p_pages_two = stats.ttest_ind(group_B['pages_viewed'],
                                       group_A['pages_viewed'],
                                       equal_var=False)
p_pages = p_pages_two / 2
print("\nPages Viewed:")
print("Mean A =", round(group_A['pages_viewed'].mean(),2),
      "Mean B =", round(group_B['pages_viewed'].mean(),2))
print("T-statistic =", round(t_pages,4))
print("One-tailed p-value =", round(p_pages,6))
```

```python
alpha = 0.05

print("\nFinal Decisions")
print("Conversion Rate:", "Significant" if p_value_conv < alpha else "Not Significant")
print("Time on Site:", "Significant" if p_time < alpha else "Not Significant")
print("Pages Viewed:", "Significant" if p_pages < alpha else "Not Significant")
```
